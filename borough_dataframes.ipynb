```python
import pandas as pd

# Assuming df1 is your original dataframe

# Extract borough names and years from the index
df1['borough'] = df1.index.str.extract(r'^(.*?)_\d{4}$', expand=False)
df1['year'] = df1.index.str.extract(r'_(\d{4})$', expand=False).astype(int)

# Drop rows where extraction failed (if any)
df1.dropna(subset=['borough', 'year'], inplace=True)

# Create a dictionary to store each borough's dataframe
borough_dfs = {}

# Loop through each unique borough and create a dataframe
for borough in df1['borough'].unique():
    borough_df = df1[df1['borough'] == borough].copy()
    borough_df.set_index('year', inplace=True)
    borough_name = borough.replace(' ', '_').replace('&', 'and')
    borough_dfs[borough_name] = borough_df

# You now have a dictionary with dataframes for each borough
# To create individual variables for each borough dataframe:
for borough_name, borough_df in borough_dfs.items():
    globals()[f'df_{borough_name}'] = borough_df

# Example: Access the dataframe for Barking_and_Dagenham
df_Barking_and_Dagenham = borough_dfs['Barking_and_Dagenham']
df_Barking_and_Dagenham
```
