```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 20/21

```python
df2021 = pd.read_csv("data/pas_data_ward_level/PAS_ward_level_FY_20_21.csv")
```

```python
questions = ['Q60', 'Q61', 'Q62A', 'Q62F', 'Q62B', 'Q62C', 'Q62TG', 'A121', 'XQ128C', 'XQ128D', 'NQ135BH', 'XQ122A', 'XQ122B']
```

```python
df_quesions = df2021[questions]
```

```python
fig, axs = plt.subplots(nrows=len(df_quesions.columns), ncols=1, figsize=(10, 5 * len(df_quesions.columns)))

for ax, column in zip(axs, df_quesions.columns):
    value_counts = df_quesions[column].value_counts()
    ax.bar(value_counts.index, value_counts.values)
    ax.set_title(f'Distribution of {column}')
    ax.set_xlabel('Categories')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
